```python
#以列表形式返回需要使用的路径下的所有所需要的文件
def fileroute(dirname,routelist):
    filelist=os.listdir(dirname);
    for i in range(len(filelist)):
        if ".log" in filelist[i]:
            routelist.append(dirname+filelist[i]);
    return routelist;
```

```python
#返回一段含有结果数值的字符串#############################################################################################################
def file(route):
    lastline=None;
    if "CCSD" in route:
        content1=open(route,"r")
        for line in content1:
            if "CORR" in line:
                lastline=line;
        content1.close();
    if "MP2" in route:
        content2=open(route,"r")
        for line in content2:
            if "EUMP2" in line:
                lastline=line;
        content2.close();
    if "B3LYP" in route:
        content3=open(route,"r")
        for line in content3:
            if "SCF Done" in line:
                lastline=line;
        content3.close();
    return lastline;
```

```python
#提取行中要素#############################################################################################################################
def extract(x):
    value = 0
    
    if "SCF Done" in x:
        # 使用正则表达式提取"E(UB3LYP)="之后到"A.U. after"之前的值
        pattern = r'E\((UB3LYP|RB3LYP)\) = (.*?) A\.U\. after'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(2).strip()
        else:
            print("未找到匹配的值")
    
    if "CORR" in x:
        # 使用正则表达式提取"E(CORR)="之后以及"Delta="之前的值
        pattern = r'E\(CORR\)=(.*?)\s+Delta=(.*?)\s+'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    if "EUMP2" in x:
        pattern = r'EUMP2\s*=\s*([-+]?\d*\.\d+([eE][-+]?\d+)?)'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    return value
```

```python
#输入路径+文件名的列表，可直接获得所需值的列表############################################################################################
#将列表中的字符串转化位np.array型字符串
#将np.array型字符串转化为np.array型浮点类型

def returnvaluelist(route,value):
    for i in range(len(route)):
        lastline=file(route[i]);
        value.append(extract(lastline));
        valuelist=np.array(value, dtype=float)
    return valuelist;
```
