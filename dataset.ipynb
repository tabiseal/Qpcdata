```python
#以列表形式返回需要使用的路径下的所有所需要的文件
#1111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111
def fileroute(dirname,routelist):
    filelist=os.listdir(dirname);
    for i in range(len(filelist)):
        if ".log" in filelist[i]:
            routelist.append(dirname+filelist[i]);
    return routelist;
```

```python
# 判断输出文件是否正常完成计算
# 获取计算非正常完成的文件列表
#22222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222222

def NormalFinishedJudge(routelist, faillist):
    for route in routelist:
        try:
            with open(route, "r") as content:
                lastline = None
                for line in content:
                    lastline = line
                if "Normal termination" not in lastline:
                    faillist.append(route)
        except FileNotFoundError:
            # 处理文件不存在的情况，可以添加适当的错误处理或日志记录
            print(f"File {route} not found")
    return faillist
```

```python

def delete_files_in_directory(path):
    try:
        # 检查路径是否存在
        if os.path.exists(path):
            # 遍历目录中的所有文件
            for root, dirs, files in os.walk(path):
                for file in files:
                    file_path = os.path.join(root, file)
                    # 删除文件
                    os.remove(file_path)
            print(f"所有文件在路径 {path} 下已删除。")
        else:
            print(f"路径 {path} 不存在。")
    except Exception as e:
        print(f"发生错误: {e}")

```

```python

def copy_file_to_directory(relative_path, target_directory):
    try:
        # 获取文件名
        file_name = os.path.basename(relative_path)
        # 构建目标路径
        target_path = os.path.join(target_directory, file_name)
        # 使用shutil.copy()函数将文件复制到目标路径
        shutil.copy(relative_path, target_path)
        print(f"文件 '{file_name}' 已成功复制到目录 '{target_directory}'")
    except Exception as e:
        print(f"复制文件时发生错误: {e}")

# 示例用法
# source_file = "example.txt"  # 输入文件的相对路径
# destination_directory = "new_directory/"  # 目标目录

# copy_file_to_directory(source_file, destination_directory)
```

```python
def file(route):
    lastline='0';
    
    content=open(route,"r");
    for line in content:
        if "CORR" in line:
            lastline=line;
        if "EUMP2" in line:
            lastline=line;
        if "SCF Done" in line:
            lastline=line;
    content.close();
    
    return lastline;
```

```python
#提取行中要素#############################################################################################################################
def extract(x):
    value = 0
    
    if "SCF Done" in x:
        # 使用正则表达式提取"E(UB3LYP)="之后到"A.U. after"之前的值
        pattern = r'E\((UB3LYP|UO3LYP|UX3LYP|UBHandHLYP|UBHandH|UCAM-B3LYP|UPBE1PBE|UHSEH1PBE|UmPW1PW91|UmPW1LYP|UmPW1PBE|UmPW3PBE|UAPFD|RB3LYP|RBHandHLYP|UBHandH|RO3LYP|RX3LYP|RCAM-B3LYP|RPBE1PBE|RHSEH1PBE|RmPW1PW91|RmPW1LYP|RmPW1PBE|RmPW3PBE|RAPFD)\) = (.*?) A\.U\. after'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(2).strip()
        else:
            print("未找到匹配的值")
    
    if "CORR" in x:
        # 使用正则表达式提取"E(CORR)="之后以及"Delta="之前的值
        pattern = r'E\(CORR\)=(.*?)\s+Delta=(.*?)\s+'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    if "EUMP2" in x:
        pattern = r'EUMP2\s*=\s*([-+]?\d*\.\d+([eE][-+]?\d+)?)'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    return value
```

```python
#输入路径+文件名的列表，可直接获得所需值的列表############################################################################################
#将列表中的字符串转化位np.array型字符串
#将np.array型字符串转化为np.array型浮点类型

def returnvaluelist(route,value):
    for i in range(len(route)):
        lastline=file(route[i]);
        value.append(extract(lastline));
        valuelist=np.array(value, dtype=float)
    return valuelist;
```

```python
def extract_strings(input_string):
    strlist = []

    # 定义正则表达式模式
    pattern = r'temp_(.*?)_(.*?)\.log(\d*\.?\d*)'

    # 使用正则表达式进行匹配
    match = re.search(pattern, input_string)

    # 如果匹配成功，则提取三个子串并添加到列表
    if match:
        first_string = match.group(1)
        strlist.append(first_string)
        second_string = match.group(2)
        strlist.append(second_string)
        third_string = match.group(3)
        strlist.append(third_string)

        return strlist
    else:
        # 如果没有匹配成功，返回包含错误信息的列表
        return ['No match found']

# 测试函数
# input_string = "./optCO(1)H(1)/temp_APFD_6-31++G(2d,2p).log1.2796747111343163"
# result = extract_strings(input_string)
# print(result)
```
