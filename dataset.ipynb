```python
#以列表形式返回需要使用的路径下的所有所需要的文件
#1111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111
def fileroute(dirname,routelist):
    filelist=os.listdir(dirname);
    for i in range(len(filelist)):
        if ".log" in filelist[i]:
            routelist.append(dirname+filelist[i]);
    return routelist;
```

```python

def delete_files_in_directory(path):
    try:
        # 检查路径是否存在
        if os.path.exists(path):
            # 遍历目录中的所有文件
            for root, dirs, files in os.walk(path):
                for file in files:
                    file_path = os.path.join(root, file)
                    # 删除文件
                    os.remove(file_path)
            print(f"所有文件在路径 {path} 下已删除。")
        else:
            print(f"路径 {path} 不存在。")
    except Exception as e:
        print(f"发生错误: {e}")

```

```python

def copy_file_to_directory(relative_path, target_directory):
    try:
        # 获取文件名
        file_name = os.path.basename(relative_path)
        # 构建目标路径
        target_path = os.path.join(target_directory, file_name)
        # 使用shutil.copy()函数将文件复制到目标路径
        shutil.copy(relative_path, target_path)
        print(f"文件 '{file_name}' 已成功复制到目录 '{target_directory}'")
    except Exception as e:
        print(f"复制文件时发生错误: {e}")

# 示例用法
# source_file = "example.txt"  # 输入文件的相对路径
# destination_directory = "new_directory/"  # 目标目录

# copy_file_to_directory(source_file, destination_directory)
```

```python
def file(route):
    lastline='0';
    
    content=open(route,"r");
    for line in content:
        if "E(CORR)" in line:
            lastline=line;
        if "EUMP2" in line:
            lastline=line;
        if "SCF Done:" in line:
            lastline=line;
    content.close();
    
    return lastline;
```

```python
#提取行中要素#############################################################################################################################
def extract(x):
    value = 0
    
    if "SCF Done:" in x:
        matches = re.search(r'-?\d+\.\d+', x)
        
        if matches:
            value=matches.group()
            
#             value = matches.group(2).strip()
        else:
            print("未找到匹配的值")
    
    if "CORR" in x:
        # 使用正则表达式提取"E(CORR)="之后以及"Delta="之前的值
        pattern = r'E\(CORR\)=(.*?)\s+Delta=(.*?)\s+'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    if "EUMP2" in x:
        pattern = r'EUMP2\s*=\s*([-+]?\d*\.\d+([eE][-+]?\d+)?)'
        matches = re.search(pattern, x)
        
        if matches:
            value = matches.group(1).strip()
        else:
            print("未找到匹配的值")
    
    return value
```

```python
#输入路径+文件名的列表，可直接获得所需值的列表############################################################################################
#将列表中的字符串转化位np.array型字符串
#将np.array型字符串转化为np.array型浮点类型

def returnvaluelist(route,value):
    for i in range(len(route)):
        lastline=file(route[i]);
        value.append(extract(lastline));
        valuelist=np.array(value, dtype=float)
    return valuelist;
```

```python
def extract_key(input_string):
    # 找到第二个 '/'
    second_slash_index = input_string.find('/', input_string.find('/') + 1)
    if second_slash_index == -1:
        return None  # 如果没有找到第二个 '/'
    
    # 找到 'temp_' 后的位置
    temp_index = input_string.find('temp_', second_slash_index)
    if temp_index == -1:
        return None  # 如果没有找到 'temp_'
    
    # 找到 '.log' 前的位置
    log_index = input_string.find('.log', temp_index)
    if log_index == -1:
        return None  # 如果没有找到 '.log'
    
    # 提取目标部分
    result = input_string[temp_index + len('temp_'):log_index]

    return result

# # 测试函数
# input_string = "./optCO(1)H(1)/temp_B97D_6-31G.log ./QST2CO(1)H(1)/temp_B97D_6-31G.log"
# result = extract_key(input_string)
# print(result)
```

```python
def printtime(route):
    content=open(route,'r');
    for line in content:
        if "Job cpu time" in line:
            content.close();
            return line;
```

```python
def extracttime(line):
    match = re.search(r'(\d+\.\d+)\s*seconds', line)
    if match:
        # 返回匹配到的数字字符串转换为浮点数
        return float(match.group(1))
    else:
        # 如果没有匹配到数字，则返回None或其他适当的值
        return None
```

```python
def find_closest_values(arr, target_value, num_values=10):
    # 计算数组中每个元素与目标值的差值
    differences = np.abs(arr - target_value)
    
    # 获取差值最小的前num_values个元素的索引
    closest_indices = np.argsort(differences)[:num_values]
    
    # 获取差值最小的前num_values个元素的值
    closest_values = arr[closest_indices]
    
    return closest_indices, closest_values
```
